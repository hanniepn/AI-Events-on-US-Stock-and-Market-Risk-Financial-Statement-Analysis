```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the path to your actual file
df = pd.read_csv('/Users/phuonganh/Downloads/fs.csv')

# Display the first few rows to confirm it's loaded correctly
print(df.head())
```

```python
sns.set(style="whitegrid")

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Companies", y="P/E")

# Customize the plot
plt.title("Distribution of P/E Ratios by Company (2016–2024)")
plt.ylabel("P/E Ratio")
plt.xlabel("Company")
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
sns.set(style="whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="Asset Turnover",
    y="ROE",
    hue="Companies",
    style="Companies",
    s=100  # marker size
)

# Plot details
plt.title("Efficiency vs Profitability (Asset Turnover vs ROE)")
plt.xlabel("Asset Turnover")
plt.ylabel("Return on Equity (ROE)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

# Define key AI event years and labels
ai_events = {
    2018: "Obama Deepfake",
    2019: "Trump AI EO",
    2020: "Alexa Launch",
    2022: "ChatGPT Launch"
}

# Create the annotated line plot for ROE
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Years", y="ROE", hue="Companies", marker="o")

# Annotate AI events
for year, label in ai_events.items():
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)
    plt.text(year + 0.1, plt.ylim()[1] * 0.95, label, rotation=90,
             verticalalignment='top', fontsize=9)

# Add titles and labels
plt.title("ROE Over Time with AI Event Annotations")
plt.xlabel("Year")
plt.ylabel("Return on Equity (ROE)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
